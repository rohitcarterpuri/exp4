# Exploratory Data Analysis - Breast Cancer Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer

# Load data
data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target

# Display basic info
print(df.info())
print(df.head())
print(df['target'].value_counts())
```

```python
# Correlation heatmap
plt.figure(figsize=(12, 10))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix[['target']].sort_values(by='target', ascending=False), 
            annot=True, cmap='coolwarm', center=0)
plt.title('Feature Correlation with Target')
plt.tight_layout()
plt.show()
```

```python
# Feature distribution
fig, axes = plt.subplots(5, 6, figsize=(15, 12))
axes = axes.flatten()

for idx, col in enumerate(df.columns[:-1]):
    axes[idx].hist(df[df['target']==0][col], alpha=0.5, label='Benign', bins=20)
    axes[idx].hist(df[df['target']==1][col], alpha=0.5, label='Malignant', bins=20)
    axes[idx].set_title(col, fontsize=8)
    axes[idx].legend()
    axes[idx].tick_params(axis='both', labelsize=6)

plt.tight_layout()
plt.show()
```
